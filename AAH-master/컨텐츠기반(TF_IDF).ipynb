```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import re
import urllib.request
from konlpy.tag import Okt
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from collections import Counter
from sklearn.model_selection import train_test_split
```

```python
f = open("/content/notuse.txt", 'r')
lines = f.readlines()
stopwords = []
for line in lines:
    line = line.replace('\n', '')
    stopwords.append(line)
f.close()  
okt = Okt()
```

```python
def preprocessing(data):
  text_data = []

  for sentence in data:
    temp_data = []
    #- 토큰화
    temp_data = okt.morphs(sentence, stem=True) 
    #- 불용어 제거
    temp_data  = [word for word in temp_data if not word in stopwords]
    text_data.append(temp_data)

  text_data = list(map(' '.join, text_data))

  return text_data
  
  
text_data = preprocessing(data['MAIN_FNCTN'])
print(text_data[0])
```

```python
column_name = ['fuction']
df = pd.DataFrame(text_data, columns = column_name)
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(max_features=405)
tfvect = vect.fit(df['fuction'])

tfidv_df = pd.DataFrame(tfvect.transform(df['fuction']).toarray(), columns = sorted(vect.vocabulary_))
tfidv_df.index = (data['PRDUCT'])
tfidv_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
item_sim = cosine_similarity(tfidv_df, tfidv_df)
```

```python
item_sim_df = pd.DataFrame(item_sim, index=(data['PRDUCT']), columns=(data['PRDUCT']))
item_sim_df.head()
```

```python
similar_index = np.argsort(-item_sim)
print(similar_index)
```
